#NOISE REDUCTION

```python
import librosa
import noisereduce as nr
import os
import numpy as np
import soundfile as sf

input_folder_path = "/content/drive/MyDrive/Fluent Speech Commands Dataset/wavs/speakers/"
output_folder_path = "/content/drive/MyDrive/Noise_reduction_output/"

os.makedirs(output_folder_path, exist_ok=True)

for folder in os.listdir(input_folder_path):
    input_path = os.path.join(input_folder_path, folder)
    output_path = os.path.join(output_folder_path, folder)
    os.makedirs(output_path, exist_ok=True)
    print(f"Folder created at: {output_path}")

    for filename in os.listdir(input_path):
        print(f"Processing file: {filename}")
        input_file_path = os.path.join(input_path, filename)

        audio, sr = librosa.load(input_file_path, sr=16000)

        reduced_noise_audio = nr.reduce_noise(y=audio, sr=sr)

        reduced_noise_audio = np.clip(reduced_noise_audio * 32767, -32768, 32767).astype(np.int16)

        output_file_path = os.path.join(output_path, filename)
        sf.write(output_file_path, reduced_noise_audio, sr, subtype='PCM_16')

    print(f"Denoised audio saved to {output_path}")
```

```python
import librosa
import noisereduce as nr
import os
import numpy as np
import soundfile as sf

input_folder_path = "/content/drive/MyDrive/Noise_reduction_output/"
output_folder_path = "/content/drive/MyDrive/Silence_removal_output"
top_db = 35

os.makedirs(output_folder_path, exist_ok=True)

for folder in os.listdir(input_folder_path):
  input_path = os.path.join(input_folder_path, folder)
  output_path = os.path.join(output_folder_path, folder)
  os.makedirs(output_path, exist_ok=True)
  print(f"Folder created at: {output_path}")
  for filename in os.listdir(input_path):
    input_file_path = os.path.join(input_path, filename)
    audio, sr = librosa.load(input_file_path, sr=None)

    non_silent_intervals = librosa.effects.split(audio, top_db=top_db)

    non_silent_audio = np.concatenate([audio[start:end] for start, end in non_silent_intervals])

    output_file_path = os.path.join(output_path, filename)
    sf.write(output_file_path, non_silent_audio, sr)

  print(f"Silence removed audio saved as: {output_path}")
```

###CREATING BATCHES

```python
import os
import shutil
import math

main_folder_path = "/content/drive/MyDrive/Silence_removal_output"
output_base_path = "/content/drive/MyDrive/Batches"

folders = [f for f in os.listdir(main_folder_path) if os.path.isdir(os.path.join(main_folder_path, f))]

batch_size = 16
num_batches = math.ceil(len(folders) / batch_size)

for batch_num in range(num_batches):
    batch_output_folder = os.path.join(output_base_path, f"batch_{batch_num + 1}")
    os.makedirs(batch_output_folder, exist_ok=True)
    print(f"Created output folder: {batch_output_folder}")

    start_index = batch_num * batch_size
    end_index = min(start_index + batch_size, len(folders))
    batch_folders = folders[start_index:end_index]

    for folder in batch_folders:
        src_folder_path = os.path.join(main_folder_path, folder)
        dest_folder_path = os.path.join(batch_output_folder, folder)
        shutil.copytree(src_folder_path, dest_folder_path)
        print(f"Copied {folder} to {batch_output_folder}")

print("All folders have been copied in batches.")
```

#LENGTH NORMALIZATION

####Batch1

```python
import librosa
import os
import numpy as np
import soundfile as sf


input_folder_path = "/content/drive/MyDrive/1_Audio_Preprocessing/Batches/batch_1"
output_folder_path = "/content/drive/MyDrive/1_Audio_Preprocessing/1.3_Length_normalized_outputs_in_batches/batch_1"
target_duration = 1.5

os.makedirs(output_folder_path, exist_ok=True)

for folder in os.listdir(input_folder_path):
    input_path = os.path.join(input_folder_path, folder)
    output_path = os.path.join(output_folder_path, folder)
    os.makedirs(output_path, exist_ok=True)
    print(f"Folder created at: {output_path}")

    for filename in os.listdir(input_path):
        input_file_path = os.path.join(input_path, filename)

        audio, sr = librosa.load(input_file_path, sr=16000)
        target_length = int(target_duration * sr)

        if len(audio) > target_length:
            audio = audio[:target_length]
        else:
            padding = target_length - len(audio)
            audio = np.pad(audio, (0, padding), 'constant')

        output_file_path = os.path.join(output_path, filename)
        sf.write(output_file_path, audio, sr, format='WAV')

    print(f"Normalized audio saved to: {output_path}")
```

```python
#subtype identification
import wave
import struct

def print_wave_info(file_path):
    with wave.open(file_path, 'rb') as wav_file:
        # Get the sample rate
        sample_rate = wav_file.getframerate()

        # Get the number of channels and sample width (bit depth)
        num_channels = wav_file.getnchannels()
        sample_width = wav_file.getsampwidth()

        # Map the sample width to the bit-depth (subtype)
        if sample_width == 1:
            bit_depth = 8
        elif sample_width == 2:
            bit_depth = 16
        elif sample_width == 3:
            bit_depth = 24
        elif sample_width == 4:
            bit_depth = 32
        else:
            bit_depth = None  # For other sample widths

        # Print the sample rate and bit-depth (subtype)
        print(f"Sample Rate: {sample_rate} Hz")
        print(f"Bit Depth (Subtype): {bit_depth} bits")
        print(f"Channels: {num_channels}")

# Example usage
file_path = "/content/drive/MyDrive/1_Audio_Preprocessing/Batches/batch_6/BvyakyrDmQfWEABb/005ed0c0-4607-11e9-bc65-55b32b211b66.wav"
print_wave_info(file_path)
```
